# Regular Expressions

- Python has a built in model named 're' that is used for Regular expressions.

```python
import re

s = "learning regex in python using Jupyter."
match = re.search(r'regex',s)
print(match)
print(match.start())
print(match.end())
print(match.span())
```

Before starting with the Python regex module let’s see how to actually write regex using metacharacters or special sequences. 

## Metacharacters

- Metacharacters are the characters with special meaning.
- List of Meta Characters: \\, [], ^, $, ., |, ?, *, +, {}, ()

## 1. \

- The backslash (\) makes sure that the character is not treated in a special way. This can be considered a way of escaping metacharacters.

```python
import re
s = "Mrs. Aishwarya Singh"
match = re.search(r'.',s)
print(match)

#The first search (re.search(r'.', s)) matches any character,
# not just the period
```

```python
import re
s = "Mrs. Aishwarya Singh"
match = re.search(r'\.',s)
print(match)

# the second search (re.search(r'\.', s)) specifically looks for
# and matches the period character.
```

## 2. []

- Square Brackets ([]) represent a character class consisting of a set of characters that we wish to match. For example, the character class [abc] will match any single a, b, or c. 
- [0, 3] is sample as [0123]
- [a-c] is same as [abc]
- We can also invert the character class using the caret(^) symbol.
- [^0-3] means any number except 0, 1, 2, or 3
- [^a-c] means any character except a, b, or c

```python
import re
s = "I am grateful for today as I am able to walk and do my chores."
pat = r"[a-m]"
result = re.findall(pat, s) 
print(result)

# In this code, you’re using regular expressions to find all 
# the characters in the string that fall within the 
# range of ‘a’ to ‘m’. The re.findall() function returns a 
# list of all such characters. In the given string, the characters 
# that match this pattern are: 'a', 'm', 'g', 'a', 'e', 'f', 'l', 'f', 'd',
# 'a', 'a', 'a', 'm', 'a', 'b', 'l', 'e', 'a', 'l', 'k', 'a', 'd', 'd', 
# 'm', 'c', 'h', 'e'
```

## 3. ^

- It matches the beginning of the string i.e. checks whether the string starts with the given character(s) or not.
- ^g will check if the string starts with g such as geeks, globe, girl, g, etc.
- ^ge will check if the string starts with ge such as geeks, geeksforgeeks, etc.

```python
import re 
regex = r'^The'
strings = ['The quick brown fox', 'The lazy dog', 'A quick brown fox'] 
for string in strings: 
    if re.match(regex, string): 
        print(f'Matched: {string}') 
    else: 
        print(f'Not matched: {string}') 
```

## 4. $

- Dollar(\$) symbol matches the end of the string i.e checks whether the string ends with the given character(s) or not. For example-
- s\$ will check for the string that ends with a such as geeks, ends, s, etc.
- ks\$ will check for the string that ends with ks such as geeks, geeksforgeeks, ks, etc.

```python
import re 
  
string = "Hello World!"
pattern = r"World!$"
  
match = re.search(pattern, string) 
if match: 
    print("Match found!") 
else: 
    print("Match not found.") 
```

## 5. .

- Dot(.) symbol matches only a single character except for the newline character (\n).
- a.b will check for the string that contains any character at the place of the dot such as acb, acbd, abbb, etc
- .. will check if the string contains at least 2 characters
Example:

```python
import re
s = "abnkc"
pat = r"a.n"
match = re.search(pat, s) 
if match: 
    print("Match found!") 
else: 
    print("Match not found.") 
```

```python
import re
s = "abnkc"
pat = r"a..k"
match = re.search(pat, s) 
if match: 
    print("Match found!") 
else: 
    print("Match not found.") 
```

## 6. |

- Or symbol works as the or operator meaning it checks whether the pattern before or after the or symbol is present in the string or not. For example –  

- a|b will match any string that contains a or b such as acd, bcd, abcd, etc.

```python
import re
s = "my name is abc"
pat = r"abc|xyz"
match = re.search(pat, s) 
if match: 
    print("Match found!") 
else: 
    print("Match not found.") 
```

## 7. ?

- matches 0 or 1 repetition
- ab?c will be matched for the string ac, acb, dabc but will not be matched for abbc because there are two b. Similarly, it will not be matched for abdc because b is not followed by c.

## 8. *

- matches 0 or more repetition
- ab*c will be matched for the string ac, abc, abbbc, dabc, etc. but will not be matched for abdc because b is not followed by c.

## 9. +

- matches 1 or more repetition
- ab+c will be matched for the string abc, abbc, dabc, but will not be matched for ac, abdc, because there is no b in ac and b, is not followed by c in abdc.

## 10. {m,n}

- matches repetition between m and n both inclusive
- a{2, 4} will be matched for the string aaab, baaaac, gaad, but will not be matched for strings like abc, bc because there is only one a or no a in both the cases.

## 11. (\<regex>) - Group

- Group symbol is used to group sub-patterns.
- (a|b)cd will match for strings like acd, abcd, gacd, etc.

## Special Sequences

- Special sequences do not match for the actual character in the string instead it tells the specific location in the search string where the match must occur. It makes it easier to write commonly used patterns.  

## 1. \w

- matches all the alphanumeric chars(A-Z, a-z, 0-9,_)

## 2. \W

- matches all except alphanumeric chars
- Matches any non-alphanumeric character.

## 3. \d

- matches all digits

## 4. \D

- except digits
- Matches any non-digit character, this is equivalent to the set class [^0-9]

## 5. \s

- matches all space chars (' ','\n','\t','\r','\v')

## 6. \S

- matches all except space characters.

## 7. \b

- matches boundary char or word
- Matches if the word begins or ends with the given character. \b(string) will check for the beginning of the word and (string)\b will check for the ending of the word.

## 8. \B

- matches the character or word if it is not the boundary word

## 9. \A

- Matches if the string begins with the given character

## 10. \Z

- Matches if the string ends with the given regex
- ab\Z : abcdab, abababab

## RegEx functions

## 1. re.findall()

- matches all the occurences in string and returns a list


```python
import re
s = """Hello my Number is 123456789 and 
            my friend's number is 987654321"""
pat = r'\d+'

match = re.findall(pat,s)
print(match)
```

## 2. re.compile

- Regular expressions are compiled into pattern objects, which have methods for various operations such as searching for pattern matches or performing string substitutions. 

```python
import re 
p = re.compile('[a-e]') 
  
print(p.findall("Aye, said Mr. Gibenson Stark")) 
```

```python
import re
s = "I went to him at 11 A.M. on 4th July 1886"
pat = r'\d'

match = re.findall(pat,s)
print(match)
```

```python
import re
s = "I went to him at 11 A.M. on 4th July 1886"
pat = re.compile('\d') 
match = re.findall(pat,s)
print(match)
```

## 3. re.split()

```python
from re import split 
  
print(split('\W+', 'Words, words , Words')) 
print(split('\W+', "Word's words Words")) 
print(split('\W+', 'On 12th Jan 2016, at 11:02 AM')) 
print(split('\d+', 'On 12th Jan 2016, at 11:02 AM')) 
print("\n")
print(re.split('\d+', 'On 12th Jan 2016, at 11:02 AM', 1)) 
print(re.split('[a-f]+', 'Aey, Boy oh boy, come here', flags=re.IGNORECASE)) 
print(re.split('[a-f]+', 'Aey, Boy oh boy, come here')) 
```

## 4. re.search()

- This method either returns None (if the pattern doesn’t match), or a re.MatchObject contains information about the matching part of the string. This method stops after the first match, so this is best suited for testing a regular expression more than extracting data.

```python
import re 
regex = r"([a-zA-Z]+) (\d+)"
  
match = re.search(regex, "I was born on June 24") 
if match != None: 
    print ("Match at index %s, %s" % (match.start(), match.end())) 
    print ("Full match: %s" % (match.group(0))) 
    print ("Month: %s" % (match.group(1))) 
    print ("Day: %s" % (match.group(2)))
else: 
    print ("The regex pattern does not match.") 
```

## 5. re.sub()

- It is used to replace a pattern from the string
- re.sub(pattern, replacement, string)

```python
s = "This is java learning."

pat = r'java'

res = re.sub(pat,'python', s)
res
```

## Sets

- A Set is a set of characters enclosed in ‘[]’ brackets. Sets are used to match a single character in the set of characters specified between brackets

## 1.\\{n,\\}

- Quantifies the preceding character or group and matches at least n occurrences.

## 2. *

- Quantifies the preceding character or group and matches zero or more occurrences.

## 3. [0123]

- Matches the specified digits (0, 1, 2, or 3)

## 4. [^arn]

- matches for any character EXCEPT a, r, and n

## 5. \d

- matches any digit(0-9)

## 6. [0-5][0-9]

- matches for any two-digit numbers from 00 and 59

## 7. \w

- Matches any alphanumeric character (a-z, A-Z, 0-9, or _).

## 8. [a-n]

- Matches any lower case alphabet between a and n.

## 9. \D

- matches any non-digit character

## 10. [arn]

- matches where one of the specified characters (a, r, or n) are present

## 11. [a-zA-Z]

- matches any character between a and z, lower case OR upper case

## 12. [0-9]

- matches any digit between 0 and 9
