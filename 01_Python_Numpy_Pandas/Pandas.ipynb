# What is Pandas?

- Pandas is a data manipulation package in Python for tabular data. 

# What is pandas used for?

- With pandas, you can:
 - Import datasets from databases, spreadsheets, comma-separated values (CSV) files, and more.
 - Clean datasets, for example, by dealing with missing values.
 - Tidy datasets by reshaping their structure into a suitable format for analysis.
 - Aggregate data by calculating summary statistics such as the mean of columns, correlation between them, and more.
 - Visualize datasets and uncover insights.

## df.loc[]

- Explicit index
- The df.loc[ ] command helps you access a group of rows and columns. loc is label-based, which means that you have to specify rows and columns based on their row and column labels.
- It includes the ends, i.e., df.loc[0:4] will return all the rows from 0-4(included).
- df.loc[start:stop:step]

```python
# selecting all the rows from 0-4 and the associated columns
data.loc[:4]
```

```python
# selecting all the rows and the column named 'Pclass'
data.loc[:,'Pclass']
```

```python
# selecting all the rows from 1-5 and the columns named 'Sex' and 'Pclass'
data.loc[:5,['Sex', 'Pclass']]
```

```python
# selecting all the rows and columns with entries having Age > 40 
data.loc[data['Age']>40.0]
```

## iloc[]

- Implicit index
- df.iloc[start:stop:step]
- The df.iloc[ ] command also helps you access a group of rows and columns. 
- Still, unlike loc, iloc is integer position-based, so you have to specify rows and columns by their integer position values.
- It excludes the ends, i.e., df.iloc[0:4] will return all the rows from 0-3 as 4 is excluded.

```python
data.iloc[:4]
```

```python
data.iloc[:,:]
```

```python
data.iloc[0:4,0:4]
```

```python
data.iloc[:5,[0,2,5]]
```

```python
data.iloc[[3,4,5],:]
```

```python
data.iat
```

## 3. dropna()

- The dropna() command drops the rows/columns with missing entries.

```python
# dropping all the rows with missing values
data_1 = data.copy()
data_1.dropna()
```

```python
# using the "axis" parameter
# "axis = 0" (default) means Row and "axis = 1" means Column
# dropping all the columns with missing entries
data_1.dropna(axis = 1)
```

```python
data_1.shape
```

```python
# using the "how" parameter
# how = "any" means dropping rows/columns having "ANY" missing entries.
# how = "all" means dropping rows/columns having "ALL" missing entries.
# dropping the columns having any missing values.
data_1.dropna(axis = 1, how = 'any')
```

```python
# using the "subset" parameter
# it is used for defining in which columns to look for missing values 
# dropping all the rows where the "duration" column is NaN
data_1.dropna(subset = ['Cabin'])
```

```python
data_1.shape
```

NOTE : The dropna() and fillna() commands returns a copy of your object rather than the actual object. To update your object , you need to specify the value of the inplace parameter as True. Donâ€™t worry, the inplace parameter is covered below.

# Checking the Duplicates

```python
name=['Hitesh','Arvind','Shipra','Jaishree','Vaibhav','Vivek','Akhilesh','Sanchita']
code=[101,102,103,104,105,106,107,108]
city=['Delhi','Mumbai','Pune','Chennai','Chandigarh','Bangalore','Noida','Gurugram']
```

```python
emp=pd.DataFrame({'Name':name,'ID':code,"City":city})
emp
```

```python
emp_comp=pd.DataFrame({'Company':['TCS','TCS','TCS','TCS','TCS','TCS','TCS','TCS']})
emp_comp
```

```python
emp_final=pd.concat((emp,emp_comp),axis=1)
emp_final
```

```python
emps = pd.DataFrame({"Name":['Vivek', 'Arvind','Shipra'],
                                "ID":[106,102,103],
                                "City":['Bangalore','Mumbai','Pune'],
                                "Company":['TCS','TCS','TCS']})
emps
```

```python
emp_finals=pd.concat((emp_final,emps),ignore_index=True)
emp_finals
```

```python
# checking for duplicates
emp_finals.duplicated()
```

```python
emp_finals.duplicated().sum()
```

# Drop the duplicates

```python
emp_finals.drop_duplicates()
```

# apply()

- It is generally used on a column and it returns an iterator object.
- Inside the apply() we use lambda() or any user defined function.

```python
# male should be converted to 1 and female should be to 0
def convert(s):
    if s=='male':
        return 1
    else:
        return 0
```
