# Matplotlib

- Matplotlib is a popular plotting library in Python used for creating high-quality visualizations and graphs. It offers various tools to generate diverse plots, facilitating data analysis, exploration, and presentation. 
- It can create different types of visualization reports like line plots, scatter plots, histograms, bar charts, pie charts, box plots, and many more different plots. This library also supports 3-dimensional plotting.

# Importing

```python
import matplotlib.pyplot as plt
import numpy as np
```

# plot()

- It is a universal function that is used to plot on x and y axes

```python
x=np.arange(1,11,0.5)
y=x**2
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

# Approaches

- 1- Functional approach
- 2- Object oriented approach

## Functional Approach

- It is the simple approach also known as beginners' approach.
- It is used to plot the data quickly.
- Functional approach we can not increase or decrease the size.

```python
# changing the line style('-', '--', '-.', ':')
plt.plot(x,y,color='red',linestyle = ':')
plt.show()
```

```python
# changing the line width
plt.plot(x,y,color='r',linestyle='-.', linewidth=3)
```

```python
# data markers('o','v','*','<','>')
plt.plot(x,y,color='red',marker='o')
plt.show()
```

```python
# size of the markers
plt.plot(x,y,color='red',marker='o',markersize=10)
plt.show()
```

```python
# Change the color of the marker
# facecolor
plt.plot(x,y,color='red',marker='o',markersize=10,markerfacecolor='green')
plt.show()
```

```python
# Change the color of the marker
# edgecolor
plt.plot(x,y,color='red',marker='o',markersize=15,markerfacecolor='yellow',markeredgecolor='blue',markeredgewidth=4)
plt.show()
```

### Legends in a plot

```python
plt.plot(x,y,color='red',marker='o',markersize=15,markerfacecolor='yellow',markeredgecolor='blue',markeredgewidth=4,label='a vs b')

plt.legend(loc=9,fontsize=20)
plt.show()
```

### Title and Axes

```python
plt.plot(x,y,color='red',marker='o',markersize=15,markerfacecolor='yellow',markeredgecolor='blue',markeredgewidth=4,label='a vs b')

plt.title('First Plot')
plt.xlabel('Array-a')
plt.ylabel('Array-b')

plt.legend(loc=9,fontsize=20)
plt.show()
```

## Object Oriented Approach

- It is also known as figure and axes approach.
- In this approach first we create a figure object then in that figure object we can create n number of axes.
- We plot the data in axes not in the figure we can plot.
- We can increase or decrease the size of the figure as well as the size of the axes.

### Create a figure object

```python
f= plt.figure()
# f is known as figure object
```

### Create axes

- figure_object.add_axes([left_boundary,bottom_boundary,width,height])

```python
f=plt.figure()

f.add_axes([0.1,0.1,0.5,0.5])
f.add_axes([0.7,0.1,0.6,0.7])
f.add_axes([0.1,0.7,0.5,0.5])
f.add_axes([0.7,0.9,0.5,0.5])

plt.show()
```

### Plot the data

- we will the plot the data in each axis

```python
f=plt.figure()

a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.6,0.7])
c=f.add_axes([0.1,0.7,0.5,0.5])
d=f.add_axes([0.7,0.9,0.5,0.5])

a.plot(x,y)
b.plot(y,x)
c.plot(x**2,y**2)
d.plot(x**2,y/2**3)

plt.show()
```

### Give title, labels

```python
f=plt.figure()

a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.6,0.7])
c=f.add_axes([0.1,0.7,0.5,0.5])
d=f.add_axes([0.7,0.9,0.5,0.5])

a.plot(x,y,label="X vs Y", marker='o')
a.set_title('Plot-a')
a.set_xlabel('x')
a.set_ylabel('y')
a.legend()

b.plot(y,x,label='Y vs X')
b.set_title('Plot-b')
b.set_xlabel('y')
b.set_ylabel('x')
b.legend()

c.plot(x**2,y**2)
c.set_title('Plot-c')

d.plot(x**2,y/2**3)
d.set_title('Plot-d')

plt.show()
```

### Increase or Decrease the size and the pixel

- plt.figure(figsize=(width,height),dpi=int)
- dpi- dots per inch (pixel density)

```python
# figsize

f=plt.figure(figsize=(12,10))

a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.6,0.7])
c=f.add_axes([0.1,0.7,0.5,0.5])
d=f.add_axes([0.7,0.9,0.5,0.5])

a.plot(x,y,label='X vs Y',marker='o')
a.set_title('Plot-a')
a.set_xlabel('x')
a.set_ylabel('y')
a.legend()

b.plot(y,x,label='Y vs X')
b.set_title('Plot-b')
b.set_xlabel('y')
b.set_ylabel('x')
b.legend()

c.plot(x**2,y**2)
c.set_title('Plot-c')

d.plot(x**2,y/2**3)
d.set_title('Plot-d')

plt.show()
```

```python
# Using DPI and save the figure
f=plt.figure(figsize=(10,6),dpi=200)

a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.5,0.5])

a.plot(x,y,color='r',marker='o',markersize=10,markerfacecolor='yellow',markeredgecolor='blue',markeredgewidth=2)
a.set_title('Plot-1',fontdict={'color':'blue'})

b.plot(y,x)
b.set_title('Plot-2')

# Save the image
# figure_object.savefig('file_name.extension')
# bbox_inches='tight'- So that the whole figure object is saved

f.savefig('demo.png',bbox_inches='tight')
plt.show()
```

# Different Types of Plot

## 1. Subplots

- Subplots() is a Matplotlib function that displays multiple plots in one figure.
- plt.subplot(rows,cols,plot_number)

```python
import numpy as np

x1 = np.linspace(0, 10, 100)

#this will create the subplots with 2 rows and 2 columns 
#and the second argument is size of the plot 
fig, ax = plt.subplots(2,2, figsize=(10,6))

# lets plot all the figures
ax[0][0].plot(x1, np.sin(x1),'g') #row=0, col=0
ax[0][1].plot(x1, np.cos(x1),'y') #row=0, col=1
ax[1][0].plot(x1, np.sin(x1),'b') #row=1, col=0
ax[1][1].plot(x1, np.cos(x1),'red') #row=1, col=1

plt.tight_layout() 
#show the plots
plt.show()
```

```python
a = [1, 2, 3, 4, 5]
b = [10, 20, 25, 30, 40]

plt.subplot(2,1,1)
plt.plot(a,b,color='red',marker='o',markerfacecolor='yellow',markeredgecolor='aqua',markersize=8,label='X vs Y')
plt.title('Plot-1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()

plt.subplot(2,1,2)
plt.plot(b,a,label='Y vs X',color='green',marker='o')
plt.title('Plot-2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()

plt.tight_layout()
plt.show()
```

## 2. Line Plots

- A line plot shows the relationship between the x and y-axis.
- The plot() function in the Matplotlib library’s Pyplot module creates a 2D hexagonal plot of x and y coordinates. plot() will take various arguments like plot(x, y, scalex, scaley, data, **kwargs).
- x, y are the horizontal and vertical axis coordinates where x values are optional, and its default value is range(len(y)).
- scalex, scaley parameters are used to autoscale the x-axis or y-axis, and its default value is actual.
- **kwargs specify the property like line label, linewidth, marker, color, etc.

```python
x1 = np.linspace(0,10,100)
plt.plot(x1, np.sin(x1),'-',color="orange")
plt.plot(x1, np.cos(x1),':',color='b')
plt.xlabel('xlabel')
plt.ylabel('ylabel')

plt.title("Title")
plt.show()
```

## 3. Histogram

- The most common graph for displaying frequency distributions is a histogram. To create a histogram, the first step is to create a bin of ranges, then distribute the whole range of values into a series of intervals and count the value that will fall in the given interval.
- It is used to plot the distribution of a discrete (numeric) column
- x: x-coordinate or sequence of the array

- bins: integer value for the number of bins wanted in the graph

- range: the lower and upper range of bins

- density: optional parameter that contains boolean values

- histtype: optional parameter used to create different types of histograms like:-bar bar stacked, step, step filled, and the default is a bar

```python
plt.figure(figsize=(10,6), dpi=100)

# default orientation is vertical
plt.hist(data.Age, bins=50, color='pink', density=True, orientation='horizontal')
plt.show()
```

## 4. Bar Plot

- Mainly, the barplot shows the relationship between the numeric and categoric values.
- In a bar chart, we have one axis representing a particular category of the columns and another axis representing the values or counts of the specific category.
- Barcharts are plotted both vertically and horizontally.
- plt.bar(x,height,width,bottom,align)
- x: representing the coordinates of the x-axis
- height: the height of the bars
- width: width of the bars. Its default value is 0.8
- bottom: It’s optional. It is a y-coordinate of the bar. Its default value is None
- align: center, edge its default value is center

## 6. Stack Plot

```python
# List of Days
days = [1, 2, 3, 4, 5]

# No of Study Hours
Studying = [7, 8, 6, 11, 7]

# No of Playing Hours
playing = [8, 5, 7, 8, 13]

# Stackplot with X, Y, colors value
plt.stackplot(days, Studying, playing,
            colors =['r', 'c','g'])

# Days
plt.xlabel('Days')

# No of hours
plt.ylabel('No of Hours')

# Title of Graph
plt.title('Representation of Study and \
Playing wrt to Days')

# Displaying Graph
plt.show()
```

## 7. Box Plot

- A box plot, also known as a box-and-whisker plot, provides a visual summary of the distribution of a dataset. It represents key statistical measures such as the median, quartiles, and potential outliers in a concise and intuitive manner.
- On the x-axis, you’ll find the data values, while the y-coordinates represent the frequency distribution.
- Parameters used in box plots are as follows:
 - data: NumPy array
 - vert: It will take boolean values, i.e., true or false, for the vertical and horizontal plot. The default is True

 - width: This will take an array and sets of the width of boxes. Optional parameters

 - Patch_artist: It is used to fill the boxes with color, and its default value is false

 - labels: Array of strings which is used to set the labels of the dataset

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
plt.boxplot(data.Age, patch_artist=True,vert=True)
plt.show()
```

## 8. Pie Chart

- A Pie Chart is a circular statistical plot that can display only one series of data.
- The area of the chart is the total percentage of the given data.
- The area of slices of the pie represents the percentage of the parts of the data.
- The slices of pie are called wedges.
- The area of the wedge is determined by the length of the arc of the wedge.
- x: Sequence of an array

- labels: List of strings which will be the name of each slice in the pie chart

- Autopct: It is used to label the wedges with numeric values. The labels will be placed inside the wedges. Its format is %1.2f%

```python
data.Sex.value_counts()
```

```python
plt.pie(data.Sex.value_counts(), labels=['male','female'])
plt.show()
```

```python
# in terms of percentage

data.Sex.value_counts()*100/len(data)

plt.pie(data.Sex.value_counts(),labels=['male','female'],autopct='%.2f%%',colors=('yellow','blue'),explode=(0.1,0.1))

plt.show()
```

```python
plt.pie(data.Pclass.value_counts(),
        labels=data.Pclass.value_counts().index,
        autopct='%.2f%%',
        colors=('pink','purple','magenta'),
        explode=(0.06,0.06,0.06))

plt.legend()
plt.show()
```

## 9. Word Cloud

- Wordcloud is the visual representation of the text data. Words are usually single, and the font size or color shows the importance of each word. The word cloud () function is used to create a word cloud in Python.
- The word cloud() will take various arguments like:
 - width: set the width of the canvas .default 400
 - height: set the height of the canvas .default 400
 - max_words: number of words allowed. Its default value is 200.
 - background_color: background color for the word-cloud image. The default color is black.

```python
#import the libraries
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
from PIL import Image 
import numpy as np
#set the figure size . 
plt.figure(figsize=(10,15)) 
#dummy text.
text = '''Donec erat diam, faucibus pulvinar eleifend vitae, vulputate quis ipsum.
          Maecenas luctus odio turpis, nec dignissim dolor aliquet id.
          Mauris eu semper risus, ut volutpat mi. Vivamus ut pellentesque sapien.
          Etiam fringilla tincidunt lectus sed interdum. Etiam vel dignissim erat.
          Curabitur placerat massa nisl, quis tristique ante mattis vitae.
          Ut volutpat augue non semper finibus. Nullam commodo dolor sit amet purus auctor'''

wordcloud = WordCloud( width=900,
    height=500,margin=0,background_color='black',
                      colormap='BuPu').generate(text) 
#imshow() function in pyplot module of matplotlib library is used to display data as an image.
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis('off') 
plt.margins(x=0, y=0) 
plt.show()
```

For instance, let’s check a line plot in a 3-D plane.

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'black')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='autumn_r');
```
